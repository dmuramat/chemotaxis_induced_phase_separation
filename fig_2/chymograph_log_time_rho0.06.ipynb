```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_dir = "../data_files/"
```

```python
def get_ordered_time_steps(file_name, group_name = 'rho/parameter_set_0'):
    file = h5py.File(file_name, 'r')
    
    group = file[group_name]
    
    list_of_time_steps = [(group[timestep_name].attrs['time'], group[timestep_name][:]) for timestep_name in group]
    
    file.close()
    
    return np.asarray([element[1] for element in sorted(list_of_time_steps)])
```

```python
sorted_time_steps = get_ordered_time_steps(data_dir + '/KSML1000rhoInit0.06_I.h5')
```

```python
plt.imshow(np.log(sorted_time_steps+1), cmap='Greys',vmin=0., vmax=.1, aspect="auto")
plt.savefig("chymograph_rho_0.03_log.png", dpi=100)
```
